```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
sd = df["review"].unique()
```

```python
oe = OrdinalEncoder(categories=[sd]) 
```

```python
dataframe = pd.DataFrame(({"size":["s","m","l","xl","s","s","l","s","m","xl", "s"]}))
```

```python
dataframe
```

```python
ord_data = [["s","m","l","xl"]]
```

```python
or_en = OrdinalEncoder(categories=ord_data)
or_en.fit(dataframe[["size"]])
```

```python
dataframe["size_en"] = or_en.transform(dataframe[["size"]])
```
