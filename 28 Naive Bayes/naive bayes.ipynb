```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv("placement.csv")
df.head(4)
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(x="cgpa", y="iq",data=df , hue="placement")
plt.show()
```

```python
x = df.iloc[: , 1 : -1]
y = df["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
sns.kdeplot(data=df["cgpa"])
plt.show()
```

```python
sns.kdeplot(data=df["iq"])
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train , y_train)
```

```python
gnb.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() ,clf=gnb)
plt.show()
```

```python
mnb = MultinomialNB()
mnb.fit(x_train , y_train)
```

```python
mnb.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() ,clf=mnb)
plt.show()
```

```python
bnb = BernoulliNB()
bnb.fit(x_train , y_train)
```

```python
bnb.score(x_test , y_test)*100
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() ,clf=bnb)
plt.show()
```
